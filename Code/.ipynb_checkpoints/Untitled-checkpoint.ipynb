```python
import pandas as pd
import numpy as np
```

```python
def calc_YTD_Avg(df,stat):
    avg=df.groupby(df.Date.dt.year)[stat].cumsum().shift()/df.groupby(df.Date.dt.year)[stat].cumcount()
    avg.name='YTD_AVG_stat'
    return avg
```
